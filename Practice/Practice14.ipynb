```python
dataframe = pd.read_csv("PracticeData/Iris Dataset.csv")
dataframe
```

```python
X = dataframe.drop("species", axis=1)
Y = dataframe["species"]
print(X.shape)
print(Y.shape)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=51)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
print(X_train)
print(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardScaler = StandardScaler()
standardScaler.fit(X_train)
X_train_sc = standardScaler.transform(X_train)
standardScaler.fit(X_test)
X_test_sc = standardScaler.transform(X_test)
```

```python
X_train_sc = pd.DataFrame(X_train_sc, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
X_test_sc = pd.DataFrame(X_test_sc, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
```
